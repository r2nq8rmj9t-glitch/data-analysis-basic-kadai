```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x1 = np.linspace(0, 10, 300)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)

data2 = np.random.randn(1000)
axes[0, 2].hist(data2)

x3 = np.random.rand(50) * 10
y3 = np.random.rand(50) * 10
axes[1, 0].scatter(x3, y3)

plt.show()
```
